## **Problem 1 - Implement Merge sort, Bubble sort, Insertion sort, Selection sort, and Quick sort on the age column of the titanic dataset**

```python
#Load Titanic Dataset
import pandas as pd
import numpy as np
gender_data = pd.read_csv('./Titanic Data Set/gender_submission.csv')
train_data = pd.read_csv('./Titanic Data Set/train.csv')
test_data = pd.read_csv('./Titanic Data Set/test.csv')

#Place ages into a list
ages = list(train_data['Age']) + list(test_data['Age'])
#Clean up ages
ages_clean = [age for age in ages if pd.notna(age)]
```

**Merge Sort**

```python
import time
```

```python
#Merge Sort
def merge(arr, start, middle, partitionSize):
    #Temp arrays for left and right half
    leftHalf = arr[start:middle+1]
    rightHalf = arr[middle+1:partitionSize+1]
    #Indexes for leftHalf, rightHalf, and arr
    left = 0
    right = 0
    i = start
    #Makes sure theres elements left in each array
    #Comapres elements from left and right and puts the smaller one into arr
    while (left < len(leftHalf) and right < len(rightHalf)):
        if leftHalf[left] <= rightHalf[right]:
            arr[i] = leftHalf[left]
            left += 1
        else:
            arr[i] = rightHalf[right]
            right += 1
        i += 1
    #Copy remaining elements from leftHalf and rightHalf
    while left < len(leftHalf):
        arr[i] = leftHalf[left]
        left += 1
        i += 1
    while right < len(rightHalf):
        arr[i] = rightHalf[right]
        right += 1
        i += 1
        
def mergeSort(arr, start, partitionSize):
    if start < partitionSize:
        middle = (start + partitionSize) // 2
        mergeSort(arr, start, middle)
        mergeSort(arr, middle + 1, partitionSize)
        merge(arr, start, middle, partitionSize)

arr = ages_clean.copy()
#Uses time to calculate how long the algorithm takes to run
start_time = time.time()
mergeSort(arr, 0, len(arr) - 1)
end_time = time.time()
print("First 20 numbers of the merge sort are: " + str(arr[:20]))
print(f"Merge sort took {end_time - start_time:.6f} seconds")
```

**Insertion Sort**

```python
def insertionSort(arr):
    #0 is already considered sorted
    for i in range(1, len(arr)):
        #Stores the current element and x is the element before the key
        key = arr[i]
        x = i - 1
        #Shifts elements greater than the key by one position to the right
        while x >= 0 and key < arr[x]:
            arr[x + 1] = arr[x]
            x -= 1
        #Places the key in the correct sorted position
        arr[x + 1] = key   
        
arr = ages_clean.copy()
#Uses time to calculate how long the algorithm takes to run
start_time = time.time()
insertionSort(arr)
end_time = time.time()
print("First 20 numbers of the insertion sort are: " + str(arr[:20]))
print(f"Insertion sort took {end_time - start_time:.6f} seconds")
```

**Selection Sort**

```python
def selectionSort(arr, length):
    for i in range(length):
        #Current position holds smallest assumed value
        smallest = i
        #Searches array for a smaller value
        for x in range(i + 1, length):
            #If a smaller value is found that becomes smallest
            if arr[x] < arr[smallest]:
                smallest = x
        #Swaps smallest value with current position
        arr[i], arr[smallest] = arr[smallest], arr[i]
arr = ages_clean.copy()
#Uses time to calculate how long the algorithm takes to run
start_time = time.time()
selectionSort(arr, len(arr))
end_time = time.time()
print("First 20 numbers of the selection sort are: " + str(arr[:20]))
print(f"Selection sort took {end_time - start_time:.6f} seconds")
```
